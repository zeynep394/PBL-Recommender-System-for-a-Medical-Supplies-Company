```python
#Cosinüs benzerliğni bulurken sadece tanım yazısı ve ürünlerin ID'si ile çalışacağız
#bu yüzden a isimi yeni bir df tanımlıyoruz.
```

```python
a=df[["Products.long_description",'Order_Items.product_id']]
a
```

```python
#bazı ID'ler birden fazla kez tekrar ettiği için 
#tekrar eden ID'leri drop ediyoruz.
a.drop_duplicates(subset ="Order_Items.product_id", 
                     keep = False, inplace = True)
a
```

```python
a.info()
#tanım sütunumuzun veri tipi obje 
#ama biz bu sütunu string olarak değerlendirmek istiyoruz bu yüzden tip değişimi uyguluyoruz
```

```python
a['Products.long_description']=a['Products.long_description'].fillna("not known")
```

```python
a['Products.long_description']=a['Products.long_description'].astype('category')
```

```python
#verileri temizleme aşaması
a['Products.long_description']=a['Products.long_description'].str.lower()
a['Products.long_description']=a['Products.long_description'].str.replace('[^\w\s]','')
a['Products.long_description']=a['Products.long_description'].str.replace('\d+','') #removing numerals 
a['Products.long_description']=a['Products.long_description'].str.replace('\n',' ').str.replace('\r','')
a
```

```python
a['Products.long_description']=a['Products.long_description'].str.replace('[^a-zA-Z]',' ')
```

```python
#apply tokenization
a['clean_text']=a['Products.long_description'].apply(lambda x: nltk.word_tokenize(x) ) 
a
```

```python
#apply stemma
from nltk.stem import PorterStemmer
stemmerporter=PorterStemmer()
import nltk
from nltk.stem import WordNetLemmatizer
lemmetizer_output= WordNetLemmatizer()
  
a['stem_text'] = a['clean_text'].apply(lambda x: [lemmetizer_output.lemmatize(j) for j in x if not j in set(STOP_WORDS)] )
a
```

```python
ylist=a['Products.long_description'].tolist()
len(ylist)
```

```python
a['cleaned_text'] = a['stem_text'].apply(lambda x: ' '.join(x) )
a
#metin verilerimiz kullanılmaya hazır.
```

```python
col_one_list=a['Products.long_description'].tolist()
col_one_list
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
corpus = col_one_list
type(corpus)
```

```python

X_train_counts = count_vect.fit_transform(corpus)

pd.DataFrame(X_train_counts.toarray(),columns=count_vect.get_feature_names(),index=[a["Products.long_description"]])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
trsfm=vectorizer.fit_transform(corpus)
pd.DataFrame(trsfm.toarray(),columns=vectorizer.get_feature_names(),index=[a["Products.long_description"]])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(trsfm[0:1], trsfm)
#cosinüs benzerliği uygula
```
